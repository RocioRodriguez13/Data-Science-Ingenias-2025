10) Selecciona el maximo de datos faltantes. Calcula que porcentaje de los datos representan los datos faltantes.

```python
# Obtengo el maximo valor de datos faltantes y lo guardo en la variable masnan
faltantes = expvida.isnull().sum()
masnam = faltantes.max()
columna_mas_faltantes = faltantes.idxmax()
porcentaje = (masnam / len(expvida)) * 100

print(f"Columna con más datos faltantes: {columna_mas_faltantes}")
print(f"Cantidad de datos faltantes: {masnam}")
print(f"Porcentaje que representan: {porcentaje:}%")
```

```python
# Obtengo la cantidad de filas total y divido el maximo valor de datos faltantes por este numero
# y lo multiplico por 100

cantidad_filas_total = len(expvida)
porcentaje_datos_faltantes = (masnam / cantidad_filas_total) * 100
```

```python
# Imprimo el maximo porcentaje de datos faltantes. Uso f-strings.

print(f"El porcentaje máximo de datos faltantes es: {porcentaje_datos_faltantes:.2f}%")
```

14) ¿Cual es la expectativa de vida mas alta y mas baja? Usa la funcion de pandas y la funcion de NumPy y compara el resultado. ¿Que paises tienen esas expectativas de vida?

```python
# Maxima expectativa de vida usando numpy

max_exp_con_np = np.max(expvida['life_expectancy'])
print(f"La máxima expectativa de vida usando NumPy es: {max_exp_con_np}")
```

```python
# Maxima expectativa de vida usando pandas
max_exp_pd = expvida['life_expectancy'].max()
print(f"La máxima expectativa de vida usando Pandas es: {max_exp_pd}")
```

```python
# Minima expectativa de vida usando numpy
minima_exp_con_np = np.min(expvida['life_expectancy'])
print(f"La mínima expectativa de vida usando NumPy es: {minima_exp_con_np}")
```

```python
# Minima expectativa de vida usando pandas
minima_exp_con_pandas = expvida['life_expectancy'].min()
print(f"La mínima expectativa de vida usando Pandas es: {minima_exp_con_pandas}")
```

```python
# Muestro los resultados maximos
print(max_exp_con_np)
print(max_exp_pd)
```

```python
# Muestro los resultados minimos
print(minima_exp_con_np)
print(minima_exp_con_pandas)
```

```python
# Filtro las columnas para mostrar solo las filas que tengan valor
# igual al maximo

filtro_columnas_max = expvida[expvida['life_expectancy'] == max_exp_con_np]
print(filtro_columnas_max)
```

```python
# Filtro las columnas para mostrar solo las filas que tengan valor
# igual al minimo

filtro_columnas_min = expvida[expvida['life_expectancy'] == minima_exp_con_np]
print(filtro_columnas_min)
```
